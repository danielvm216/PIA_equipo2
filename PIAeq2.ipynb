```python
import time #importamos time para establecer tiempos entre ejecuciones
from datetime import datetime #importamos datetime para hacer operaciones con fechas

#creamos una uncion que nos permita calcular edad fertil restante para las mujeres de la familia
def calcular_puntos_fertilidad(edad):
    if 20 <= edad <= 35:
        return 35 - edad
    elif edad < 20:
        return 15
    else:
        return 0

#creamos una funcion que nos permita hacer operaciones con fechas de nacimiento y obtener la edad actual
def obtener_fecha_nacimiento():
    fecha_nacimiento_str = input("Ingresa la fecha de nacimiento en el siguiente formato (DD/MM/AAAA): ")
    fecha_nacimiento = datetime.strptime(fecha_nacimiento_str, "%d/%m/%Y")
    return fecha_nacimiento

#creamos una variable a la cual le asignamos el valor de la fecha al momento en que se está ejecutando el codigo
fecha_actual = datetime.now()

#creamos una lista con los 10 oficios solicitados en Australia
oficios = ["1- Carpintero", "2- Constructor", "3- Agricultor", "4- Enfermero", "5- Barbero", "6- Pintor", "7- Electricista", "8- Chofer", "9- Taxista" , "10- Secretario"]


# inicializamos nuestra variable acumuladora de puntaje
puntaje_total = 0

print("Bienvenido al programa de Inmigrantes a Australia") #bienvenida al programa
print("-"*50)
time.sleep(1.5) # damos un tiempo de espera de 1.5 segundos para seguir ejecutando la bienvenida
print("Vamos a comenzar a contabilizar tu puntaje")
time.sleep(1) # damos un tiempo de espera de 1 segundo

# obtenemos si la familia es de uno o dos integrantes, y, con la ayuda de la funcion upper, nos evitamos errores de sensibilidad a mayusculas
tipo_familia = input("¿La familia es monoparental (M) o biparental (B)? ").upper()


#creamos una condicion que solo se ejecutara si la familia es monoparental
if tipo_familia == "M":
    sexo_padre1 = input("Ingresa el sexo del progenitor 1 (M/F/N): ").upper()
    if sexo_padre1 == "F": # Creamos una condicion en la que, obtendremos la edad fertil restante en caso de que el progenitor sea de sexo femenino
        puntaje_total += 6
        fecha_nacimientoF = obtener_fecha_nacimiento() #mandamos llamar la funcion que nos hace el calculo y nos devuelve la edad en datatype int
        edad_F = (fecha_actual - fecha_nacimientoF).days // 365
        puntaje_total += calcular_puntos_fertilidad(edad_F) # añadimos a nuestra variable acumuladora de puntos el resultado de restar 35 (edad fertil maxima) - la edad de la progenitora
    #en caso de que el progenitor sea de sexo masculino, no hacemos los calculos para obtener la edad fertil restante
    else:
        fecha_nacimiento_padre1 = obtener_fecha_nacimiento()
        puntaje_total += 6 #como es de sexo masculino, solo agregamos los 6 puntos correspondientes


    #comenzamos a preguntar grado academico del progenitor
    print("Captura la letra correspondiente al grado académico del progenitor:")
    print(" P: Postgraduado; U: Universitario; T: Técnico; N: Ninguno.")
    grado_academico_progenitor = input("").upper()
    if grado_academico_progenitor == "P": #en caso de que el progenitor tenga un postgrado, añadimos 5 puntos
        puntaje_total += 5
    elif grado_academico_progenitor == "U": #en caso de que el progenitor tenga un postgrado, añadimos 3 puntos
        puntaje_total += 3

    # avisamos que se va a desplegar una lista con 10 oficios
    print("A continuación se mostrará una lista con los 10 oficios requeridos")
    time.sleep(0.5)

    #mostramos los 10 elementos de la lista, con un tiempo de espera de 0.2 segundos entre cada elemento de la misma
    for i in oficios:
        time.sleep(0.2)
        print(i)


    #preguntamos si el progenitor cuenta con alguno de los oficios antes mostrados y, si es el caso, añadir 8 puntos al puntaje final
    respuesta_oficio = input("¿El progenitor cuenta con algún oficio mencionado? (S/N): ").upper()
    if respuesta_oficio == "S":
        puntaje_total += 8


#en caso de que no sea monoparental, preguntamos datos de ambos progenitores, repitiendo el proceso anterior por cada progenitor
else:
    sexo_padre1 = input("Ingresa el sexo del progenitor 1 (M/F/N): ").upper()
    #en caso de que un progenitor sea de sexo femenino, calculamos edad fertil restante
    if sexo_padre1 == "F":
        puntaje_total += 6
        fecha_nacimientoF = obtener_fecha_nacimiento()
        edad_F = (fecha_actual - fecha_nacimientoF).days // 365
        puntaje_total += calcular_puntos_fertilidad(edad_F)

    #si el progenitor es de sexo masculino solo añadimos los 6 puntos correspondientes
    else:
        fecha_nacimiento_padre1 = obtener_fecha_nacimiento()
        puntaje_total += 6

    #se pregunta el nivel de estudios de cada progenitor, y dado sea el caso, añadir su respectivo puntaje
    print("Captura la letra correspondiente al grado académico del progenitor 1:")
    print(" P: Postgraduado; U: Universitario; T: Técnico; N: Ninguno.")
    grado_academico_progenitor = input("").upper()
    if grado_academico_progenitor == "P":
        puntaje_total += 5
    elif grado_academico_progenitor == "U":
        puntaje_total += 3

    sexo_padre2 = input("Ingresa el sexo del progenitor 2 (M/F/N): ").upper()
    if sexo_padre2 == "F":
        puntaje_total += 6
        fecha_nacimientoF = obtener_fecha_nacimiento()
        edad_F = (fecha_actual - fecha_nacimientoF).days // 365
        puntaje_total += calcular_puntos_fertilidad(edad_F)
    else:
        fecha_nacimiento_padre2 = obtener_fecha_nacimiento()
        puntaje_total += 6

    print("Captura la letra correspondiente al grado académico del progenitor 2:")
    print(" P: Postgraduado; U: Universitario; T: Técnico; N: Ninguno.")
    grado_academico_progenitor = input("").upper()
    if grado_academico_progenitor == "P":
        puntaje_total += 5
    elif grado_academico_progenitor == "U":
        puntaje_total += 3

    #mostramos la lista con los 10 oficios requeridos
    print("A continuación se mostrará una lista con los 10 oficios requeridos")
    time.sleep(0.5)

    #mostramos la lsita de oficios dejando un tiempo de 0.2 segundos entre cada uno de ellos
    for i in oficios:
        time.sleep(0.2)
        print(i)

    #se pregunta si el progenitor cuenta con alguno de los oficios mostrados y, si es el caso, sumarle su respectivo puntaje
    oficio_progenitor1 = input("¿El progenitor 1 cuenta con algún oficio mencionado? (S/N): ").upper()
    if oficio_progenitor1 == "S":
        puntaje_total += 8

    oficio_progenitor2 = input("¿El progenitor 2 cuenta con algún oficio mencionado? (S/N): ").upper()
    if oficio_progenitor2 == "S":
        puntaje_total += 8

# Preguntar por la cantidad de hijos
cantidad_hijos = int(input("Ingresa la cantidad de hijos que hay en la familia: "))
puntaje_hijos = cantidad_hijos * 8
puntaje_total += puntaje_hijos

# Ciclo para ingresar datos de cada hijo
for _ in range(cantidad_hijos):
    sexo_hijo = input("Ingresa el sexo del hijo (M/F/N): ").upper()

    #si se ingresa un hijo de sexo femenino, reutilizamos la funcion que usamos para determinar la edad fertil de la madre, pero ahora para determinar la edad fertil de la hija
    if sexo_hijo == "F":
        fecha_nacimiento_hijo = obtener_fecha_nacimiento()
        edad_hijo = (fecha_actual - fecha_nacimiento_hijo).days // 365
        puntaje_total += calcular_puntos_fertilidad(edad_hijo)

    #en caso de que se ingrese un hijo de sexo masculino, se omiten los calculos y solo se pregunta la fecha de nacimiento
    else:
        fecha_nacimiento_hijo = obtener_fecha_nacimiento()

# Mostrar el puntaje total
print("El puntaje total es de: ", puntaje_total)

```
